```python
#Q7 Visualizations
# sklearn is used only for quickly splitting the data into a train and test set
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
from DecisionTree import DecisionTree
import numpy as np
big_dset = np.loadtxt("data\Dbig.txt")
train_8192, test_set = train_test_split(big_dset, train_size=0.8192)
training_sets = [train_8192]
for i in range(4):
    training_sets.append(train_test_split(training_sets[i], train_size=0.25)[0])
    print(training_sets[i + 1].shape[0])


def class_error(model:DecisionTree, test_set):
    classified_test = np.array([[test[0], test[1], test[2], model.classify_point(test)] for test in test_set])
    mis_classification = np.sum(np.abs(classified_test[:, 2] - classified_test[:, 3]))
    # Error function 
    return mis_classification/test_set.shape[0]

dt_stats = list()
for train_set in training_sets:
    decision_tree = DecisionTree(train_set)
    decision_tree.train_dt()
    decision_tree.plot_dt()
    plt.title(f'{train_set.shape[0]} Training Points')
    plt.show()
    dt_error = class_error(decision_tree, test_set)
    dt_nodes = decision_tree.trained_dt.count_nodes()
    dt_stats.append([train_set.shape[0], dt_error, dt_nodes])
dt_df = pd.DataFrame(dt_stats, columns=["Data_Size", "DT_Error_Rate", "DT_Nodes"]).sort_values(by="Data_Size")
dt_df.plot.line(x="DT_Nodes", y="DT_Error_Rate")
plt.title("DT Learning Curve")
print(dt_df)
```

```python
# SKLearn section 
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
sk_dt_stats = list()
for train_set in training_sets:
    classifier = DecisionTreeClassifier()
    classifier.fit(train_set[:, :2], train_set[:,2])
    predictions = classifier.predict(test_set[:, :2])
    error = np.sum(np.abs(predictions - test_set[:, 2]))/ test_set.shape[0]
    count = classifier.tree_.node_count
    sk_dt_stats.append([train_set.shape[0], error, count])

skdt_df = pd.DataFrame(sk_dt_stats, columns=["Data_Size", "DT_Error_Rate", "DT_Nodes"]).sort_values(by="Data_Size")
print(skdt_df)
skdt_df.plot.line(x="DT_Nodes", y="DT_Error_Rate")
plt.title("SKlearn DT Learning Curve")
plt.show()
```

```python
# Lagrange interpolation section 
from scipy.interpolate import lagrange
import numpy as np
rng = np.random.default_rng()
# Used 17 points since more than 20 points is not reccomended for scipy.interpolate.lagrange()
x_test_vals = np.sort(np.random.uniform(0, 2*np.pi, size=17))
y_test_vals = np.sin(x_test_vals)

x_train_vals = np.sort(np.random.uniform(0, 2* np.pi, size=17))
y_train_vals = np.sin(x_train_vals)

noise_stdev = [0, 0.01, 0.05, 0.2, 0.5]
results = list()
for noise in noise_stdev:
    if noise == 0: 
        train_noise_array = np.zeros(17)
        test_noise_array = np.zeros(17)
    else:
        train_noise_array = rng.normal(0, noise, size=17)
        test_noise_array = rng.normal(0, noise, size=17)
    # Add noise to train and test X
    test_x_final = x_test_vals + test_noise_array
    train_x_final = x_train_vals + train_noise_array

    # Fit data, make predictions and calculate RMSE 
    fit = lagrange(train_x_final, y_train_vals)
    predicted_y_test = fit(test_x_final)
    predicted_y_train = fit(train_x_final)
    rmse_test = np.sqrt(np.mean((predicted_y_test - y_test_vals)**2))
    rmse_train = np.sqrt(np.mean((predicted_y_train- y_train_vals)**2))
    results.append([noise, rmse_train, rmse_test, fit.coefficients.shape[0]])
    plt.plot(train_x_final, y_train_vals)
    plt.title(f'Noise STDev = {noise}')
    plt.show()
result_df = pd.DataFrame(results, columns=['noise_stdev', "rmse_train", 'rmse_test', 'degree'])
print(result_df)
```
